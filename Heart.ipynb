```python
x = data[['age','cp','slp','thalachh','O2']]
y = data['output']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
model_linear = LinearRegression().fit(xtest,ytest)
```

```python
round(100*model_linear.score(xtest,ytest),2)
```

```python
model_dt = DecisionTreeClassifier().fit(xtest,ytest)
```

```python
round(100*model_dt.score(xtest,ytest),2)
```

```python
model_dt.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
ypred = model_dt.predict(xtest)
```

```python
print(mean_absolute_error(ytest,ypred))
```
